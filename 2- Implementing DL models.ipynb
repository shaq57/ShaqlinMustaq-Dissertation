```python
import pandas as Sls_Fcst_BuIm_pd
import matplotlib.pyplot as Sls_Fcst_BuIm_mplb
import seaborn as Sls_Fcst_BuIm_sbn
import numpy as Sls_Fcst_BuIm_num
Sls_Fcst_BuIm = Sls_Fcst_BuIm_pd.read_csv('Quan U Furniture.csv')
Sls_Fcst_BuIm.shape
```

```python
## Dataset Partitioning
```

```python
def SalesForecast_database(S, N):
  Sls_Fcst_BuIm_X = []
  Sls_Fcst_BuIm_Y = []
  for z in range(len(S)):
    rest = z + N
    if rest > len(S)-1:
      break
    rest_i, rest_o = S[z:rest], S[rest]
    Sls_Fcst_BuIm_X.append(rest_i)
    Sls_Fcst_BuIm_Y.append(rest_o)
  return Sls_Fcst_BuIm_num.array(Sls_Fcst_BuIm_X), Sls_Fcst_BuIm_num.array(Sls_Fcst_BuIm_Y)
```

```python
### Training the dataset
N = 2
Sls_Fcst_BuIm_X_TR, Sls_Fcst_BuIm_Y_TR = SalesForecast_database(Sls_Fcst_BuIm['Furniture Sales'][:-30].tolist(), N)
### Testing the dataset
Sls_Fcst_BuIm_X_TS, Sls_Fcst_BuIm_Y_TS = SalesForecast_database(Sls_Fcst_BuIm['Furniture Sales'][-30:].tolist(), N)
for k in range(len(Sls_Fcst_BuIm_X_TS)):
  print(Sls_Fcst_BuIm_X_TS[k], Sls_Fcst_BuIm_Y_TS[k])
```

```python
# Reshaping the dataset
Sls_Fcst_BuIm_X_TR =Sls_Fcst_BuIm_X_TR.reshape(Sls_Fcst_BuIm_X_TR.shape[0],Sls_Fcst_BuIm_X_TR.shape[1] , 1)
display("training input re_shaped ")
display(Sls_Fcst_BuIm_X_TR.shape)

Sls_Fcst_BuIm_X_TR =Sls_Fcst_BuIm_X_TS.reshape(Sls_Fcst_BuIm_X_TS.shape[0],Sls_Fcst_BuIm_X_TS.shape[1] , 1)
display("testing input re_shaped ")
display(Sls_Fcst_BuIm_X_TS.shape)
```

```python
## Implementing Deep Learning - LSTM model
```

```python
from tensorflow.keras.layers import LSTM as Sls_Fcst_BuIm_LSTM
from tensorflow.keras.models import Sequential as Sls_Fcst_BuIm_Sqtl
from tensorflow.keras.layers import Dense as Sls_Fcst_BuIm_Dse
from sklearn.metrics import r2_score as Sls_Fcst_BuIm_rse
from sklearn.metrics import mean_absolute_error as Sls_Fcst_BuIm_mae
from sklearn.metrics import mean_squared_error as Sls_Fcst_BuIm_mse
import math as Sls_Fcst_BuIm_mth
```

```python
Sls_Fcst_BuIm_Atvn = 'elu'
Sls_Fcst_BuIm_lss = 'mean_squared_error'
```

```python
for Sls_Fcst_BuIm_Opz in ['adam', 'adagrad', 'rmsprop']:

  print('Result = Changing optimizer as', Sls_Fcst_BuIm_Opz)

  from tensorflow.keras.layers import LSTM as Histo_pals

  Sls_Fcst_BuIm_mdl = Sls_Fcst_BuIm_Sqtl()
  Sls_Fcst_BuIm_mdl.add(Sls_Fcst_BuIm_LSTM(100, input_shape=(None,1), activation=Sls_Fcst_BuIm_Atvn))
  Sls_Fcst_BuIm_mdl.add(Sls_Fcst_BuIm_Dse(1))

  Sls_Fcst_BuIm_mdl.compile(loss= Sls_Fcst_BuIm_lss, optimizer= Sls_Fcst_BuIm_Opz)

  Sls_Fcst_BuIm_mod = Sls_Fcst_BuIm_mdl.fit(Sls_Fcst_BuIm_X_TR, Sls_Fcst_BuIm_X_TR, epochs=50, batch_size=100)

  Sls_Fcst_BuIm_Pre = Sls_Fcst_BuIm_mdl.predict(Sls_Fcst_BuIm_X_TS)
  print(Sls_Fcst_BuIm_Pre)

  print("  R2 ", Sls_Fcst_BuIm_rse(Sls_Fcst_BuIm_Y_TS, Sls_Fcst_BuIm_Pre)*100)
  print("  MAE ", Sls_Fcst_BuIm_mae(Sls_Fcst_BuIm_Y_TS, Sls_Fcst_BuIm_Pre))
  print("  MSE ", Sls_Fcst_BuIm_mse(Sls_Fcst_BuIm_Y_TS, Sls_Fcst_BuIm_Pre))
  print("  RMSE ", Sls_Fcst_BuIm_mth.sqrt(Sls_Fcst_BuIm_mse(Sls_Fcst_BuIm_Y_TS, Sls_Fcst_BuIm_Pre)))

  Furte_Stck = (Sls_Fcst_BuIm['Date'][-30:-2].tolist())
  Stck = Sls_Fcst_BuIm_pd.DataFrame()
  Stck['Sls_Fcst_BuIm_Y_TS']=Sls_Fcst_BuIm_Y_TS
  Stck['Sls_Fcst_BuIm_Pre']=Sls_Fcst_BuIm_Pre
  Stck['Furte_Stck']=Furte_Stck
  Stck=Stck.set_index("Furte_Stck")

  Sls_Fcst_BuIm_mplb.figure(figsize=(15, 6))
  Sls_Fcst_BuIm_mplb.plot(Stck["Sls_Fcst_BuIm_Pre"], label="predd", color='red')
  Sls_Fcst_BuIm_mplb.plot(Stck["Sls_Fcst_BuIm_Y_TS"], label="real", color='blue')
  Sls_Fcst_BuIm_mplb.xticks(rotation=90)
  Sls_Fcst_BuIm_mplb.title(' Real vs Prediction')
  Sls_Fcst_BuIm_mplb.xlabel('Date')
  Sls_Fcst_BuIm_mplb.ylabel('Furniture Sales')
  Sls_Fcst_BuIm_mplb.show()
```
